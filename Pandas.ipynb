# merge_asof

In principle, this function is similar to a standard left join, however, we match on the nearest key rather than equal keys. 

- backward ; for each row in the left DataFrame we select the last row in the right DataFrame whose 'on' key is less than or equal to the left's key.


- forward ; for each row in the left DataFrame we select the first row in the right Dataframe whose 'on' key is greater than or equal to the left's key.


- nearest ; for each row in the left DataFrame we search selects the row in the right DataFrame whose 'on' key is closet in absolute distance to the left's key.





```python
trades = pd.DataFrame(
       {
           "time": [
               pd.Timestamp("2016-05-25 13:30:00.023"),
               pd.Timestamp("2016-05-25 13:30:00.038"),
               pd.Timestamp("2016-05-25 13:30:00.048"),
               pd.Timestamp("2016-05-25 13:30:00.048"),
               pd.Timestamp("2016-05-25 13:30:00.048")
           ],
           "ticker": ["MSFT", "MSFT", "GOOG", "GOOG", "AAPL"],
           "price": [51.95, 51.95, 720.77, 720.92, 98.0],
           "quantity": [75, 155, 100, 100, 100]
       }
   )

quotes = pd.DataFrame(
    {
        "time": [
            pd.Timestamp("2016-05-25 13:30:00.023"),
            pd.Timestamp("2016-05-25 13:30:00.023"),
            pd.Timestamp("2016-05-25 13:30:00.030"),
            pd.Timestamp("2016-05-25 13:30:00.041"),
            pd.Timestamp("2016-05-25 13:30:00.048"),
            pd.Timestamp("2016-05-25 13:30:00.049"),
            pd.Timestamp("2016-05-25 13:30:00.072"),
            pd.Timestamp("2016-05-25 13:30:00.075")
        ],
        "ticker": [
               "GOOG",
               "MSFT",
               "MSFT",
               "MSFT",
               "GOOG",
               "AAPL",
               "GOOG",
               "MSFT"
           ],
           "bid": [720.50, 51.95, 51.97, 51.99, 720.50, 97.99, 720.50, 52.01],
           "ask": [720.93, 51.96, 51.98, 52.00, 720.93, 98.01, 720.88, 52.03]
    }
)
```

```python
df = pd.merge_asof(
    trades, quotes, on='time', by='ticker', tolerance=pd.Timedelta('2ms')
)
df
```

```python
df2 = pd.merge_asof(
    trades, 
    quotes, 
    on='time', 
    by='ticker', 
    tolerance=pd.Timedelta('2ms'),
    direction ='forward'
)
df2
```

```python
df3 = pd.merge_asof(
    trades, 
    quotes, 
    on='time', 
    by='ticker', 
    tolerance=pd.Timedelta('2ms'),
    direction ='nearest'
)
df3
```
